```python
df.head()
print(df.shape)

# Filtro para el régimen estacionario
datos_filtrado=df[(df['TIME'] % 600==540)]

print('tiempo')
print(datos_filtrado.shape)

#Reseteo el índice y quito la primera fila
datos_filtrado = datos_filtrado.reset_index(drop=True)
print(datos_filtrado.iloc[0]['TIME'])
print(datos_filtrado.iloc[1]['TIME'])

datos_filtrado=datos_filtrado.drop(0)
print(datos_filtrado.iloc[0]['TIME'], '\n')

print('wd')
#Obtener la dirección wd mímina y máxima de datos_filtrado
min=datos_filtrado['wd'].min()
max=datos_filtrado['wd'].max()
print(min, max)

datos_filtrado = datos_filtrado[datos_filtrado['wd'] >= 0]
datos_filtrado = datos_filtrado[datos_filtrado['wd'] < 360]
print(datos_filtrado.shape)
min=datos_filtrado['wd'].min()
max=datos_filtrado['wd'].max()
print(min, max, '\n')

print('WF_Power')
# Obtener el mínimo y máximo de potencia
min=datos_filtrado['WF_Power'].min()
max=datos_filtrado['WF_Power'].max()
print(min, max)

# datos_filtrado = datos_filtrado[datos_filtrado['WF_Power'] >= 0]
datos_filtrado[datos_filtrado['WF_Power'] < 0] = 0  # Mejores resultados
print(datos_filtrado.shape, '\n')

print('ws')
#Obtener el mínimo y máximo de velocidad de viento ws
min=datos_filtrado['ws'].min()
max=datos_filtrado['ws'].max()
print(min, max)

# datos_filtrado = datos_filtrado[datos_filtrado['ws'] >= 0]
datos_filtrado[datos_filtrado['ws'] < 0] = 0  # Mejores resultados
print(datos_filtrado.shape, '\n')

datos_filtrado.head()
```

```python
min_pow=datos_filtrado['WF_Power_SP'].min()
max_pow=datos_filtrado['WF_Power_SP'].max()
print(min_pow, max_pow)

salto=500000

setpoints=range(int(min_pow), int(max_pow), salto)

# setpoints=datos_filtrado['WF_Power_SP'].unique()
print(setpoints)
print(len(setpoints))
print('type: ', type(setpoints))

```

```python
def obtener_direcciones(df_filtrado, columnas):

  direcciones={}
  for i in range(0,360):
      # Separo los datos para cada dirección
      filtered_data = df_filtrado[(df_filtrado['wd'] <= i) & (df_filtrado['wd'] > (i - 1 ))]

      # Selecciono solo las columnas que me interesan
      selected_columns = filtered_data[columnas]

      direcciones[i]=selected_columns

  return direcciones
```

# Gráficas interactivas


```python
"""Gráfica de viento y potencia para cada setpoint diferenciado por direcciones y sin ellas"""


app = Dash('app1')

app.layout = html.Div([

  # dcc.Dropdown(setpoints_ordenados, setpoints_ordenados[0], id='potencia', clearable=False),
  html.Div([
      "potencia: ",
      dcc.Input(id='potencia', min=10000000, step=salto, value=10000000, type='number')
  ]),

  dcc.Checklist(
          id='direccion',
          options=[{'label': 'Mostrar datos por dirección:', 'value': 'ver_dir'}],
          value=[]
      ),
    html.Div([
      "Direccion: ",
      dcc.Input(id='dir_1', value=258, type='number')
  ]),
  dcc.Graph(id='graph')
])


@app.callback(
    Output('graph', 'figure'),
    Input('dir_1', 'value'),
    Input('direccion', 'value'),
    Input('potencia', 'value'))
def update_output_div(dir1, direccion, potencia):

  datos_limpios = datos_filtrado[(datos_filtrado['WF_Power_SP'] >= potencia) & (datos_filtrado['WF_Power_SP'] < potencia+salto)]

 # Crear figura base
  fig = go.Figure()

  if 'ver_dir' in direccion:
    direcciones=obtener_direcciones(datos_limpios, ['ws', 'WF_Power', 'wd'])
    if dir1 in direcciones:
        fig.add_trace(go.Scatter(x=direcciones[dir1]['ws'], y=direcciones[dir1]['WF_Power'], mode='markers', marker=dict(color='blue')))

  else:

    fig.add_trace(go.Scatter(x=datos_limpios['ws'], y=datos_limpios['WF_Power'], mode='markers', marker=dict(color='blue')))

  fig.update_layout(
      title='Potencia vs velocidad general y por dirección',
      transition_duration=10,
      xaxis_title='Velocidad del viento (ws)',
      yaxis_title='Potencia del parque (WF_Power)'
  )

  return fig



if __name__ == '__main__':
    app.run_server(debug=True, port=8061)
```

```python
"""Histograma de número de direcciones por cada setpoint"""

app = Dash('app1')

app.layout = html.Div([
    # dcc.Dropdown(setpoints_ordenados, setpoints[0], id='potencia', clearable=False),
    html.Div([
      "potencia: ",
      dcc.Input(id='potencia', min=10000000, step=salto, value=10000000, type='number')
  ]),
    dcc.Graph(id='graph')
])


@app.callback(
    Output('graph', 'figure'),
    Input('potencia', 'value'))
def update_output_div( potencia):

  datos_limpios = datos_filtrado[(datos_filtrado['WF_Power_SP'] >= potencia) & (datos_filtrado['WF_Power_SP'] < potencia+salto)]


  # Crear figura base
  fig = go.Figure()

  fig.add_histogram(
    x=datos_limpios['wd'],
    nbinsx=360,
    xbins=dict(start=0, end=360, size=1)
)

  fig.update_layout(
      title='Número de datos de cada dirección',
      transition_duration=10,
      xaxis_title='Direcciones de viento (wd)',
      yaxis_title='Número de veces que aparece'
  )

  return fig



if __name__ == '__main__':
    app.run_server(debug=True, port=8062)
```

```python
"""Histograma de número de velocidades por cada setpoint"""

app = Dash('app1')

app.layout = html.Div([
    # dcc.Dropdown(setpoints, setpoints[0], id='potencia', clearable=False),
    html.Div([
      "potencia: ",
      dcc.Input(id='potencia', min=10000000, step=salto, value=10000000, type='number')
  ]),
    dcc.Graph(id='graph')
])


@app.callback(
    Output('graph', 'figure'),
    Input('potencia', 'value'))
def update_output_div( potencia):

  datos_limpios = datos_filtrado[(datos_filtrado['WF_Power_SP'] >= potencia) & (datos_filtrado['WF_Power_SP'] < potencia+salto)]


  # Crear figura base
  fig = go.Figure()

  fig.add_histogram(
    x=datos_limpios['ws'],
  )

  fig.update_layout(
      title='Número de datos de cada velocidad',
      transition_duration=10,
      xaxis_title='Velocidades de viento (ws)',
      yaxis_title='Número de veces que aparece',
      xaxis_range=[0, 30]
  )

  return fig



if __name__ == '__main__':
    app.run_server(debug=True, port=8063)
```
