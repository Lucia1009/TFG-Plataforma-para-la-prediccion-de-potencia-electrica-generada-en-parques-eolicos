```python
# Concatenate the dataframes
reshaped=np.array([df_sin_5['WF_Power'],  df5_5['WF_Power'],  df10_5['WF_Power'],
                   df_sin_10['WF_Power'], df5_10['WF_Power'], df10_10['WF_Power'],
                   df_sin_15['WF_Power'], df5_15['WF_Power'], df10_15['WF_Power'],
                   df_sin_20['WF_Power'], df5_20['WF_Power'], df10_20['WF_Power']])
df_combined = pd.DataFrame(reshaped)
df_combined = df_combined.T
df_combined.columns = ['ws_5_d_0', 'ws_5_d_5', 'ws_5_d_10', 'ws_10_d_0', 'ws_10_d_5', 'ws_10_d_10', 'ws_15_d_0', 'ws_15_d_5', 'ws_15_d_10', 'ws_20_d_0', 'ws_20_d_5', 'ws_20_d_10']

# Save the combined dataframe to a CSV file
df_combined.to_csv('potencias_sin_saltos.csv', index=False)

df_combined.head()



```

## Gráficas

```python
# Inicializar la aplicación Dash
app = Dash(__name__)

# Layout de la aplicación
app.layout = html.Div([
    html.Label("Velocidad de viento (ws):"),
    dcc.Dropdown(
        id='ws',
        options=[{'label': f"WS {ws}", 'value': ws} for ws in [5, 10, 15, 20]],
        value=5,
        clearable=False
    ),
    html.Label("Ángulo de desorientación (yaw_ref):"),
    dcc.Dropdown(
        id='d',
        options=[{'label': f"D {d}", 'value': d} for d in [0, 5, 10]],
        value=0,
        clearable=False
    ),
    dcc.Graph(id='graph')
])

# Callback para actualizar la gráfica
@app.callback(
    Output('graph', 'figure'),
    Input('ws', 'value'),
    Input('d', 'value')
)
def update_graph(ws, d):

    fig = go.Figure()
    # Generar el nombre de la columna basado en ws y d
    columna = f"ws_{ws}_d_{d}"


    fig = px.line_polar( r=df_combined[columna], theta=range(0,360), line_close=True)

    fig.update_layout(
            transition_duration=10,
            width=600,
            height=600,
            xaxis_title='Velocidad del viento (ws)',
            yaxis_title='Diferencia'
        )

    return fig

# Ejecutar la aplicación
if __name__ == '__main__':
    app.run_server(debug=True, port=8060)
```

```python
# Inicializar la aplicación Dash
app = Dash(__name__)

# Layout de la aplicación
app.layout = html.Div([
    html.Label("Velocidad de viento (ws):"),
    dcc.Dropdown(
        id='ws',
        options=[{'label': f"WS {ws}", 'value': ws} for ws in [5, 10, 15, 20]],
        value=5,
        clearable=False
    ),
    html.Label("Ángulo de desorientación (yaw_ref):"),
    dcc.Dropdown(
        id='d',
        options=[{'label': f"D {d}", 'value': d} for d in [0, 5, 10]],
        value=0,
        clearable=False
    ),
    dcc.Graph(id='graph')
])

# Callback para actualizar la gráfica
@app.callback(
    Output('graph', 'figure'),
    Input('ws', 'value'),
    Input('d', 'value')
)
def update_graph(ws, d):

    fig = go.Figure()
    # Generar el nombre de la columna basado en ws y d
    columna = f"ws_{ws}_d_{d}"

    fig = fig.add_trace(go.Scatter(x=list(range(0,360)), y=df_combined[columna], marker=dict(color='blue')))

    fig.update_layout(
            transition_duration=10,
            xaxis_title='Dirección',
            yaxis_title='Potencia'
        )

    return fig

# Ejecutar la aplicación
if __name__ == '__main__':
    app.run_server(debug=True, port=8061)
```
